Here, I will define the hypothesis,

$$ h_\theta(x) = \theta^T x = \theta_0 + \theta_1 x_1 $$

and the cost function,

$$ J(\theta) = \frac 1 {2m} \sum\limits_{i=1}^m (h_\theta(X^{(i)}) - y^{(i)})^2 $$

and gradient descent:

simultaneously update $ \theta_j $ for all $j$ do
$$ \{ \theta_j := \theta_j - \alpha \frac 1 m \sum\limits_{i=1}^m (h_\theta(X^{(i)}) - y^{(i)}) x_j^{(i)} \} $$

```python
# just library code
def dot(a, b, *args, **kwargs):
    # m x n * n x o
    if len(a.shape) == 1:
        if a.shape[0] == b.shape[0]:
            a = a[newaxis, :]
        elif 1 == b.shape[0]:
            a = a[:, newaxis]
        else:
            raise ValueError("Matrix sizes don't match")
    if len(b.shape) == 1:
        if b.shape[0] == a.shape[1]:
            b = b[:, newaxis]
        elif 1 == a.shape[1]:
            b = b[newaxis, :]
        else:
            raise ValueError("Matrix sizes don't match")
    r = np.dot(a, b)
    if r.shape[0] == 1:
        return r[0, :]
    elif r.shape[1] == 1:
        return r[:, 0]
    else:
        return r
```

```python
X = vstack((ones((m)), data[0])) # Add a column of ones to x
theta = zeros(2) # initialize fitting parameters
iterations = 1500
alpha = 0.01

def hypothesis(theta, X):
    return dot(theta.T, X)

def cost(theta, X, y):
    return 1.0 / (2.0 * m) * sum((hypothesis(theta, X) - y)**2)

def iterate(iterations, theta, alpha, X, y):
    hist = []
    for i in range(iterations):
        theta = theta - alpha / m * sum((hypothesis(theta, X) - y)[newaxis, :] * X, axis=1)
        hist.append(theta)
    return theta, hist

print cost(theta, X, y)
```

I can verify that my cost function is correct, since it gives 32.07, as expected. The math has an almost 1 to 1 correspondence with the python code.

So the normal equation finds basically the same results as the gradient descent, except without iterating. The normal equation, $ (X^T X)^{-1} X^T y $). According to [this stackoverflow post](http://stackoverflow.com/questions/18191890/why-gradient-descent-when-we-can-solve-linear-regression-analytically), the normal equation works best with few features and data-points (less than 1000 or 10000), whereas the gradient descent is much faster for many features and datapoints.
