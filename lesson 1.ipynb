Here, I will define the hypothesis,

$$ h_\theta(x) = \theta^T x = \theta_0 + \theta_1 x_1 $$

and the cost function,

$$ J(\theta) = \frac 1 {2m} \sum\limits_{i=1}^m (h_\theta(X^{(i)}) - y^{(i)})^2 $$

and gradient descent:
simultaneously update $ \theta_j $ for all $j$ do {
$ \theta_j := \theta_j - \alpha \frac 1 m \sum\limits_{i=1}^m (h_\theta(X^{(i)}) - y^{(i)}) x_j^{(i)} $ }

```python
X = vstack((ones(m), data[:, 0])) # Add a column of ones to x
theta = zeros((2, 1)) # initialize fitting parameters
iterations = 1500;
alpha = 0.01;

def hypothesis(x, theta):
    return dot(theta.T, x)

def cost(theta):
    return 1.0 / (2.0 * m) * sum((hypothesis(X, theta)[0] - y)**2)

def iterate():
    global theta
    theta2 = theta.copy()
    for j in range(len(theta)):
        theta2[j] = theta[j] - alpha / m * sum((hypothesis(X, theta)[0] - y) * X[j])
    theta = theta2

print cost(theta)
```

I can verify that my cost function is correct, since it gives 32.07, as expected. The math has an almost 1 to 1 correspondence with the python code.

```python
def tostring(ndarray):
    return ','.join(map(str, map(functools.partial(round, ndigits=2), map(float, list(ndarray)))))

print 'First iteration {0}'.format(tostring(theta))
iterate()
print 'Second iteration {0}'.format(tostring(theta))
for x in range(iterations): iterate()
print 'After {iterations} iteartions {0}'.format(tostring(theta), **locals())
```
